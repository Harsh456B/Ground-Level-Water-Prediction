```python
x = df.iloc[:, [5, 6, 7, 10, 11, 12]].values
y = df.iloc[:, 13].values
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
### pie chart representation for net annum ground water

import matplotlib.pyplot as plt

# Sum up the net groundwater availability per state
statewise_ground_water = df.groupby('Name of State')['Net Ground Water Availability for future use'].sum()

# Sort states by availability and select the top 10; group the rest as "Others"
top_states = statewise_ground_water.nlargest(10)
other_states_sum = statewise_ground_water[~statewise_ground_water.index.isin(top_states.index)].sum()
top_states['Others'] = other_states_sum

# Extract labels and sizes
labels = top_states.index
sizes = top_states.values
colors = plt.cm.Paired(range(len(labels)))  # Using a colormap for clear distinction

# Plotting the pie chart
plt.figure(figsize=(6, 12))
plt.pie(sizes, colors=colors, autopct='%1.1f%%', startangle=90)
plt.legend(labels, loc="best")  # Place legend on the side
plt.title("Net Ground Water Availability for Future Use by State (Top 10 + Others)")
plt.axis('equal')  # Ensure the pie chart is circular
plt.show()
```

### Predicting through random forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Correctly selecting the feature and target columns
x = df.iloc[:, [5, 6, 7, 10, 11, 12]].values  # Use a list directly
y = df.iloc[:, 13].values

# Encode target variable if it's categorical
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)
# Handle missing values in the features using mean imputation
imputer = SimpleImputer(strategy='mean')
x = imputer.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.55,random_state=0)
```

```python
# Train a Random Forest Classifier
rf_classifier = RandomForestClassifier()
rf_classifier.fit(x_train, y_train)
```

```python
# Print the accuracy on the training and testing sets using the Random Forest classifier
training_accuracy = rf_classifier.score(x_train, y_train)
testing_accuracy = rf_classifier.score(x_test, y_test)

print("Accuracy on training set: %f" % training_accuracy)
print("Accuracy on testing set: %f" % testing_accuracy)
```

```python
random=RandomForestClassifier()
random.fit(x_train,y_train)

print("Accuracy on training set: %f" % training_accuracy)
print("Accuracy on testing set: %f" % testing_accuracy)

rf_result=random.score(x_test,y_test)
print("Random Forest:",rf_result)
```

```python
import pandas as pd
import pickle
import numpy as np

# Step 1: Load the uploaded dataset
input_file_path = "Dynamic_2017_2_0.csv"  # Replace with the actual path
data = pd.read_csv(input_file_path)

# Step 2: Preprocess the dataset
# Remove non-numeric columns and rows with missing values
data_cleaned = data.drop(columns=["S.no.", "Name of State", "Name of District"], errors="ignore")
data_cleaned.dropna(inplace=True)

# Keep only the features required for prediction (remove the target column if it exists)
if "Stage of Ground Water Extraction (%)" in data_cleaned.columns:
    prediction_data = data_cleaned.drop(columns=["Stage of Ground Water Extraction (%)"])
else:
    prediction_data = data_cleaned

# Step 3: Load the saved model and feature names
model_path = "random_forest_model.pkl"  # Replace with actual model path
features_path = "features.pkl"          # Replace with actual feature file path

with open(model_path, "rb") as f:
    loaded_model = pickle.load(f)

with open(features_path, "rb") as f:
    feature_names = pickle.load(f)

# Validate that the input data matches the features expected by the model
if list(prediction_data.columns) != feature_names:
    raise ValueError("Input data does not match the model's expected features!")

# Step 4: Predict using the loaded model
predictions = loaded_model.predict(prediction_data)

# Step 5: Add predictions to the dataset
data_cleaned["Predicted Stage of Ground Water Extraction (%)"] = predictions

# Save the updated dataset with predictions
output_file_path = "predicted_groundwater_data.csv"  # Replace with desired output file path
data_cleaned.to_csv(output_file_path, index=False)

# Step 6: Display the predictions
print("Predictions:")
print(data_cleaned[["Predicted Stage of Ground Water Extraction (%)"]].head())
```
